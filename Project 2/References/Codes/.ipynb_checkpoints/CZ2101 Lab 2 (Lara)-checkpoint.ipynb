### Priority Queue (Array)

```python
#Implementation of Priority Queue using a queue

class PriorityQueue(object):
    
    def __init__(self):
        self.queue = []
        
    def __str__(self):
        return ' '.join([str(i) for i in self.queue])
    
    def length(self):
        return len(self.queue)     
    
    # insert element into queue
    def insert(self, num):
        self.queue.append(num)
        
    # check for empty queue
    def isEmptyQueue(self):
        return len(self.queue) == 0
        
    # pop element with highest priority
    def pop(self):
        
        min = -1
        
        for i in range(len(self.queue)):
            
            if self.queue[i][0] < self.queue[min][0]:
                min = i
                
            element = self.queue[min]
            del self.queue[min]
            
            return element
        
    def remove(self, temp):
        
        for i in range(len(self.queue)):
            
            if self.queue[i][1] == temp:
                
                del self.queue[i]
                break
```

### Dijkstra Algorithm - Adjacent Matrix, Array Priority Queue

```python
def dijkstra_adj_matrix(g, source):
    
    d = [math.inf for i in range(len(g))]
    d[0] = 0
    pi = [-1 for j in range(len(g))]
    
    S = [0 for i in range(len(g))]
    
    pq = PriorityQueue()
    
    for v in range(len(g)):
        pq.insert([d[v], v])
        
    while not pq.isEmpty():
        
        #obtain cheapest node
        u = pq.pop()
        S[u[1]] = 1
        
        for v in range(len(g[u[1], :])):
            
            if g[u[1]][v] != 0:
                
                if (S[v] != 1 and d[v] > d[u[1]] + g[u[1]][v]):
                    
                    for j in range(pq.length()):
                        
                        if(pq.queue[j][1] == v):
                            
                            pq.remove(j)
                            break
                    
                    #update distance
                    d[v] = d[u[1]] + g[u[1]][v] 
                    
                    #update predecessor
                    pi[v] = u[1]
                    
                    #insert v into priority queue according to its updated distance
                    pq.insert([d[v], v])
    
    return d
```

### Dijkstra Algorithm - Adjacency Lists, Minimising Heap Priority Queue

```python
def dijkstra_adj_lists(g, source):        
                    
                    
                    
    d = [math.inf for i in range(len(g))]
    d[source] = 0
    pi = [-1 for i in range(len(g))]
    
    S = [0 for i in range(len(g))]
    
    pq = [(d[v], v) for v in range(len(g))]
    
    #minimising heap priority queue
    heapq.heapify(pq)
    
    while(pq):
        #obtain cheapest node
        u = heapq.heappop(pq)
        
        S[u[1]] = 1
        
        for v in g[u[1]]:
            
            if (S[v[0]] != 1 and d[v[0]] > d[u[1]] + v[1]):
                
                for i in range(len(pq)):
                    
                    if (pq[i][1] == v):
                        pq.pop(i)
                        
                #update distance    
                d[v[0]] = d[u[1]] + v[1]
                    
                #update predecessor
                pi[v[0]] = u[1]
                    
                #insert v into priority queue according to its updated distance
                heapq.heappush(pq, (d[v[0]], v[0]))
    
    return d
```
